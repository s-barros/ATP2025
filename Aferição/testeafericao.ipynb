```python
#a) Contar ocorrências: defina uma função que recebe uma lista e um elemento como argumento e devolve o número de elementos na lista que são maiores que o elemento passado como argumento.

def contaOcur(lista,elem):
    if lista == []:
        return 0
    if lista[0]>elem:
        num = 1
    else:
        num = 0
    return num + contaOcur(lista[1:],elem)
    
print(contaOcur([],2))
print(contaOcur([1,2,3,4,5],2))
```

```python
#b) Disjuntas?: defina uma função que recebe duas listas como argumento e devolve o valor booleano verdadeiro se as listas não tiverem nenhum elemento comum e falso caso contrário.

def disjuntas(lista1,lista2):
    if lista1 == []:
        return True
    else:
        if lista1[0] in lista2:
            return False 
        if lista1[0] not in lista2:
            return disjuntas(lista1[1:],lista2)

print(disjuntas([1,2,3],[4,5,6]))
print(disjuntas([1,2,3],[3,4,5]))
```

```python
# mergeListas: defina uma função que recebe duas listas de números ordenadas crescentemente como argumento e devolve uma nova lista ordenada de forma crescente com os elementos das duas listas.

def mergeListas(lista1,lista2):
    if lista1 == []:
        return lista2
    else:
        return sorted([lista1[0]] + mergeListas(lista1[1:],lista2))
    
print(mergeListas([1,2,3],[4,5,6]))
print(mergeListas([1,2,3],[2,3,4,5,6]))
```

```python
#a) splitLista: defina uma função que recebe dois argumentos, uma lista de números e um número, e dá como resultado um tuplo composto por duas listas, a primeira contendo os números divisíveis pelo número passado como argumento, e a segunda contendo os números restantes.

def splitLista(lista,num):
    divisiveis = []
    ndivisiveis = []
    for elem in lista:
        if elem%num==0:
            divisiveis.append(elem)
        else:
            ndivisiveis.append(elem)
    return (divisiveis,ndivisiveis)

print(splitLista([1,2,3,4,5], 2))
```

```python
#mmc: defina uma função que calcula o mínimo múltiplo comum entre dois números inteiros.

def mmc(num1,num2):
    i = 1
    mul1 = num1
    while mul1 % num2 != 0:
        i = i + 1
        mul1 = num1*i
    return mul1

print(mmc(12, 16))
```

```python
#c) max2prod: defina uma função que recebe uma lista de números inteiros positivos e devolve o maior produto que for possível calcular multiplicando os 2 maiores inteiros da lista.

def max2prod(lista):
    lista.sort()
    return lista[-1]*lista[-2]

print(max2prod([5,3,2,4,1]))
```

```python
#d) reduxInt: defina uma função que recebe um número inteiro positivo como argumento e vai adicionar repetidamente os seu dígitos até obter apenas um dígito que é retornado como resultado.

def reduxInt(n):
    while len(str(n)) != 1:
        soma = 0
        for num in str(n):
            soma = soma + int(num)
        n = soma
    return n

print(reduxInt(998))
```

```python
#a) listar_unidades: Defina uma função que recebe um hospital como argumento e devolve a lista dos nomes das unidades hospitalares desse hospital

hospital = [
 (
 "Unidade de Cardiologia",
 [ (1, "Dr. João"), (2, "Dra. Maria")],
 [(101, False), (102, True), (103, False), (104, False)]
 ),
 (
 "Unidade de Pediatria",
 [ (3, "Dr. Pedro"), (4, "Dra. Clara") ],
 [ (201, True), (202, True), (203, True)]
 )
]

def listar_unidades(hosp):
    res = []
    for uni,_,_ in hosp:
        res.append(uni)
    return res

print(listar_unidades(hospital))
```

```python
#b) camas_disponiveis: Defina uma função que recebe um hospital como argumento e devolve o número total de camas disponiveis nesse hospital.

def camas_disponiveis(hosp):
    num = 0
    for _,_,camas in hosp:
        for cama in camas:
            if cama[1]==False:
                num = num + 1
    return num 

print(camas_disponiveis(hospital))
```

```python
#c) listar_medicos_unidade: Defina uma função que recebe um hospital e o nome de uma unidade hospitalar como argumento e devolve a lista dos médicos dessa unidade.

def listar_medicos_unidade(hosp,uni):
    for u,m,_ in hosp:
        if u == uni:
            return m 
        
print(listar_medicos_unidade(hospital, "Unidade de Cardiologia"))
```

```python
#d) existe_medico: Defina uma função que recebe um hospital, o nome de uma unidade hospitalar e o nome de um médico, e verifica se o médico existe nessa unidade hospitalar, devolvendo True caso exista e False em caso contrário.

def existe_medico(hosp,uni,med):
    existe = False
    for u,m,_ in hosp:
        if u == uni:
            for medico in m:
                if medico[1] == med:
                    existe = True
    return existe

print(existe_medico(hospital, "Unidade de Cardiologia", "Dra. Maria"))
```

```python
#e) adicionar_medico: Defina uma função que recebe um hospital, o nome de uma unidade hospitalar e o nome de um médico, e caso este não exista na unidade, acrescenta-o e devolve o novo hospital com o médico inserido.

def adicionar_medico(hosp,uni,med):
    existe = False
    for u,m,_ in hosp:
        if u == uni:
            for medico in m:
                if medico == med:
                    existe = True
            if existe == False:
                m.append((len(m)+1,med))
    return hosp

print(adicionar_medico(hospital, "Unidade de Cardiologia", "Dra. Paula"))
            
```

```python
#f) ocupar_cama: Defina uma função que recebe um hospital, o nome de uma unidade hospitalar e o identificador de uma cama, e caso a cama esteja livre devolve um novo hospital com a cama ocupada (não faz nada se a cama estiver ocupada).

def ocupar_cama(hosp,uni,idc):
    for u,_,c in hosp:
        if u == uni:
            for cama in c:
                if cama[0]==idc:
                    if cama[1]==False:
                        c.remove(cama)
                        c.append((idc,True))
    return hosp

print(ocupar_cama(hospital, "Unidade de Cardiologia", 101))
```

```python
#g) guardar_hospital: Defina uma função que recebe um hospital e o nome de um ficheiro e guarda a informação do hospital no ficheiro (o formato do ficheiro fica ao seu critério).

def guardar_hospita(hosp,ficheiro):
    f = f.open(ficheiro,"w",encoding="utf-8")
    f.write(hosp)
    f.close()
```

```python
#h) carregar_hospital: Defina uma função que recebe o nome de um ficheiro e carrega a informação do ficheiro para uma variável em memória que é devolvida como resultado (o formato do ficheiro deverá ser o mesmo que foi usado na alínea anterior).

def carregar_hospital(ficheiro):
    res = []
    f = f.open(ficheiro,"r",encoding="utf-8")
    for elem in f:
        res.append(elem)
    f.close()
    return res
```

```python
#i) camas_ocupadas: Defina uma função que recebe um hospital e devolve a percentagem de camas ocupadas do total de camas no hospital.

def camas_ocupadas(hosp):
    numerototal = 0
    numeroocupadas = 0
    for _,_,c in hosp:
        for cama in c:
            numerototal += 1 
            if cama[1] == True:
                numeroocupadas += 1
    return (numeroocupadas/numerototal)*100
```

```python
#j) unidades_ocupacao_critica: Defina uma função que recebe um hospital e devolve uma lista de tuplos (pares), em que o primeiro elemento do tuplo é o nome de uma unidade hospitalar e o segundo é a percentagem de camas ocupadas nessa unidade, mas apenas deverão constar tuplos em que a percentagem é maior ou igual a 80%

def unidades_ocupacao_critica(hosp):
    res = []
    for u,_,c in hosp:
        camas = 0
        camasoc = 0
        for cama in c:
            camas += 1
            if cama[1]== True:
                camasoc += 1
        if (camasoc/camas)*100 >= 80:
            res.append((u,(camasoc/camas)*100))
    return res
```
